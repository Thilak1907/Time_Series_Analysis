```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Example: Load your time series dataset (or use any time series data)
# For demonstration, we will simulate a simple time series with trend and seasonality.

# Create a time series with trend and seasonality for demonstration
np.random.seed(0)
time = pd.date_range(start='2020-01-01', periods=100, freq='D')
trend = np.linspace(10, 50, 100)  # Simulated trend component
seasonality = 10 * np.sin(np.linspace(0, 3.14 * 2, 100))  # Simulated seasonal component
noise = np.random.normal(0, 5, 100)  # Random noise

# Combine them to create a time series data with trend + seasonality + noise
data = trend + seasonality + noise
df = pd.DataFrame({'date': time, 'value': data})
df.set_index('date', inplace=True)

# Plot the original time series data
plt.figure(figsize=(10, 6))
plt.plot(df['value'], label='Original Time Series')
plt.title('Original Time Series Data')
plt.legend()
plt.show()

# Decompose the time series data into trend and seasonality using seasonal_decompose
result = seasonal_decompose(df['value'], model='additive', period=30)  # period is set to 30 for seasonality

# Plot only the trend and seasonal components
plt.figure(figsize=(12, 8))

# Plot the trend component
plt.subplot(211)
plt.plot(result.trend, label='Trend', color='green')
plt.legend(loc='upper left')
plt.title('Trend')

# Plot the seasonal component
plt.subplot(212)
plt.plot(result.seasonal, label='Seasonal', color='orange')
plt.legend(loc='upper left')
plt.title('Seasonal')

plt.tight_layout()
plt.show()
```
